```python
import pandas as pd
```

```python
tanks = pd.read_csv("../data/assets/tanks.csv")
```

```python
tanks
```

```python
# Label encoding "type" column
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(tanks.type.unique())
tanks["type_le"] = le.transform(tanks.type)
tanks
```

```python
# If tank has no turret set hull armor as turret armor
tanks.armor_turret = tanks.armor_turret.fillna(tanks.armor_hull)
tanks = tanks.convert_dtypes()
tanks
```

```python
# Clustering function
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

def cluster(n, lvl=None):
    if lvl is None:
        filter_ = [True] * len(tanks)
    else:
        filter_ = tanks.lvl == lvl
    y_pred = KMeans(n_clusters=n, random_state=42).fit_predict(
        StandardScaler().fit_transform(
            tanks[filter_].drop(
                ["wg_id", "wr_id", "tag", "name", "short_name", "type"], axis=1)
        )
    )
    return pd.Series(data=[i if lvl is None else f"l{lvl}_{i}" for i in y_pred], index=tanks[filter_].tag)\
        .reset_index().rename(columns={0:"cluster"})
```

```python
from sklearn.manifold import TSNE
import plotly.express as px

def plot_clusters(pred, lvl=None):
    if lvl is None:
        filter_ = [True] * len(tanks)
    else:
        filter_ = tanks.lvl == lvl

    vects = StandardScaler().fit_transform(
        tanks[filter_].drop(
            ["wg_id", "wr_id", "tag", "name", "short_name", "type"], axis=1)
    )

    vects_tsne = TSNE(n_components=2, verbose=50).fit_transform(vects, axis=1)

    fig = px.scatter(
        x=vects_tsne[:, 0],
        y=vects_tsne[:, 1],
        color=list(pred),
        hover_data=[tanks[filter_].name, pred,
                    tanks[filter_].level, tanks[filter_].type]
    )

    fig.show()
```
